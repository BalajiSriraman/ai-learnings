```python
#pipline for custom prog
from datasets import load_dataset
from transformers import pipeline
from rich import print

# Function to load and preprocess data
def load_and_preprocess_data(dataset_name="xsum", split="train", num_samples=5): #2
    try:
        # Load dataset
        dataset = load_dataset(dataset_name, split=split)
        # Return a small sample for quick testing
        return dataset[:num_samples]  # Modify to return a small sample
    except Exception as e:
        print(f"[bold red]Error loading dataset: {e}[/bold red]")
        return None

# Function to create the summarizer pipeline
def create_summarizer(model_name="t5-small"): #3
    try:
        summarizer = pipeline(task="summarization", model=model_name) #pipeline
        return summarizer
    except Exception as e:
        print(f"[bold red]Error creating summarizer pipeline: {e}[/bold red]")
        return None

# Function to summarize text using the summarizer pipeline
def summarize_text(text, summarizer, min_length=20, max_length=60): #4
    try:
        result = summarizer(text, min_length=min_length, max_length=max_length, truncation=True)
        return result[0]["summary_text"]
    except Exception as e:
        print(f"[bold red]Error summarizing text: {e}[/bold red]")
        return None
# Main function to orchestrate the process
def main(): #1
    # Load the dataset
    xsum_sample = load_and_preprocess_data() #2 returns dataset

    if not xsum_sample:
        return  # If the dataset couldn't be loaded, exit the program

    # Create the summarizer pipeline
    summarizer = create_summarizer() #3 returns pipeline output

    if not summarizer:
        return  # If the summarizer couldn't be created, exit the program

    # Access the first document correctly
    document = xsum_sample["document"][0]
    print(f"[bold blue]Original Document:[/bold blue]\n{document}\n")

    # Summarize the document
    summary = summarize_text(document, summarizer) #4 return summary

    if summary:
        print(f"[bold green]Summary:[/bold green]\n{summary}")
    else:
        print("[bold red]Failed to generate summary[/bold red]")

# Run the main function
if __name__ == "__main__":
    main() #1

```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from datasets import load_dataset
from rich import print

# Load dataset
xsum_dataset = load_dataset("xsum", version="1.2.0")
xsum_sample = xsum_dataset["train"].select(range(10))  # Select a sample of 10 documents

# Load pre-trained tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

# Function to summarize articles
def summarize_articles(articles):
    # Preprocess articles by adding the "summarize:" prefix
    formatted_articles = list(map(lambda article: "summarize: " + article, articles))

    # Tokenize the input
    inputs = tokenizer(formatted_articles, max_length=1024, return_tensors="pt", padding=True, truncation=True)

    # Generate summaries
    summary_ids = model.generate(
        inputs.input_ids,
        attention_mask=inputs.attention_mask,
        num_beams=2,
        min_length=0,
        max_length=60,
    )

    # Decode the generated summaries
    decoded_summaries = tokenizer.batch_decode(summary_ids, skip_special_tokens=True)
    return decoded_summaries

# Get the document texts from the sample dataset
articles = xsum_sample["document"]

# Automate summarization for all articles
summaries = summarize_articles(articles)

# Print each summary
for i, summary in enumerate(summaries):
    print(f"[bold blue]Original Article {i + 1}:[/bold blue]\n{articles[i]}\n")
    print(f"[bold green]Summary {i + 1}:[/bold green]\n{summary}\n")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd
from rich import print

# Example: Load a product descriptions dataset (could be from a CSV, a database, etc.)
data = {
    "product_id": [1, 2, 3],
    "product_description": [
        "This is a high-performance laptop with a 16GB RAM, 512GB SSD, and an Intel i7 processor. Perfect for gaming and heavy tasks.",
        "An elegant wristwatch with a leather strap and stainless steel body. Features water resistance and a stylish design.",
        "A wireless mouse with ergonomic design, responsive clicks, and long battery life. Perfect for office use and casual browsing."
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Load the pre-trained tokenizer and model.
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

# Summarization: Prepend "summarize: " to each product description.
articles = list(map(lambda description: "summarize: " + description, df["product_description"]))

# Tokenize the input
inputs = tokenizer(
    articles, max_length=1024,
    return_tensors="pt", padding=True, truncation=True
)

# Generate summaries
summary_ids = model.generate(
    inputs.input_ids,
    attention_mask=inputs.attention_mask,
    num_beams=2,
    min_length=0,
    max_length=60,
)

# Decode the generated summaries
decoded_summaries = tokenizer.batch_decode(summary_ids, skip_special_tokens=True)

# Print summaries
for i, summary in enumerate(decoded_summaries):
    print(f"Product {df['product_id'][i]} Summary: {summary}")
```

* Zero-Shot Classification with DeBERTa-V3
*  This program classifies text into predefined categories without any prior training on them.

*  Process
*  Define a text query (e.g., a customer looking for a smartphone).
*  Provide candidate labels (Technology, Fashion, Food).
*  Use the DeBERTa-v3 model to predict the most relevant category.
*  Output the predicted label and its confidence score.
*  Key Takeaways
*  Zero-shot learning allows classification without fine-tuning on specific labels.
* DeBERTa-v3 uses Natural Language Inference (NLI) to determine the best category match.

```python
from transformers import pipeline

def zero_shot_classification(text, candidate_labels):
    try:
        # Initialize the zero-shot classification pipeline
        zero_shot_pipeline = pipeline(
            task="zero-shot-classification",
            model="cross-encoder/nli-deberta-v3-small",
            use_fast=False,
        )

        # Run zero-shot classification
        result = zero_shot_pipeline(text, candidate_labels)

        # Print the results in a more detailed format
        print("Text:", text)
        print("\nPredicted Labels and Confidence Scores:")
        for label, score in zip(result['labels'], result['scores']):
            print(f"Label: {label}, Confidence Score: {score:.4f}")

        # Return the best label and score for further use if needed
        return result['labels'][0], result['scores'][0]

    except Exception as e:
        print(f"An error occurred: {e}")
        return None, None

# Example usage
text = "I am looking for a new smartphone. I want something with a great camera and long battery life."
candidate_labels = ["Technology", "Fashion", "Food"]

best_label, confidence_score = zero_shot_classification(text, candidate_labels)

# Optionally print the best result
if best_label:
    print(f"\nBest Predicted Label: {best_label}")
    print(f"Confidence Score: {confidence_score:.4f}")
```

Steps for Few-Shot Learning with GPT-Neo for Sentiment Classification
* Load the pre-trained GPT-Neo model using Hugging Face’s pipeline for text generation.

* Define the end-of-sequence (EOS) token ("###") to structure the prompt.

* Create a prompt with labeled examples, providing a few sentiment-labeled restaurant reviews.
* Append a new review to the prompt, leaving the [Sentiment]: blank for model prediction.

* Generate text using the model, with max_new_tokens=10 to control output length.

* Extract sentiment from the model's response using regex to match the [Sentiment]: format.
* Print the final sentiment classification for the new review.

```python
from transformers import pipeline
import re

# Initialize the pipeline
few_shot_pipeline = pipeline(
    task="text-generation",
    model="EleutherAI/gpt-neo-1.3B",
    max_new_tokens=10,
)

# Define the EOS token ID
eos_token_id = few_shot_pipeline.tokenizer.encode("###")[0]

# Define the initial few-shot prompt
prompt = """
For each restaurant review, describe its sentiment:
[Review]: "The food at this restaurant was absolutely delicious, and the service was excellent."
[Sentiment]: Positive
###
[Review]: "I had a terrible experience at this place. The food was cold, and the staff was rude."
[Sentiment]: Negative
###
[Review]: "The ambiance and decor were charming, but the food quality was mediocre."
[Sentiment]: Neutral"""

# Define the new restaurant review
new_review = "The atmosphere of this restaurant was pleasant, but the food was disappointing. I wouldn't recommend it."

# Update the prompt with the new restaurant review
updated_prompt = prompt + f"\n[Review]: \"{new_review}\"\n[Sentiment]:"

# Generate text based on the updated prompt using few-shot learning and EOS token ID
try:
    results = few_shot_pipeline(updated_prompt, eos_token_id=eos_token_id)

    # Extract the sentiment label from the results for the new review
    if results and len(results) > 0:
        generated_text = results[0]["generated_text"].strip()

        # Use regex to capture the sentiment
        pattern = r"\[Sentiment\]:\s*(\w+)"
        matches = re.findall(pattern, generated_text)

        if matches:
            last_sentiment = matches[-1].lower()  # Get the last match and convert to lowercase
            print(f"Predicted Sentiment: {last_sentiment}")
        else:
            print("Sentiment not found in the generated text.")
    else:
        print("Error: No results returned from the model.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from transformers import pipeline
import re

# Initialize the pipeline
few_shot_pipeline = pipeline(
    task="text-generation",
    model="EleutherAI/gpt-neo-1.3B",
    max_new_tokens=10,
)

# Define the EOS token ID
eos_token_id = few_shot_pipeline.tokenizer.encode("###")[0]

# Define the initial few-shot prompt
prompt = """
For each restaurant review, describe its sentiment:
[Review]: "The food at this restaurant was absolutely delicious, and the service was excellent."
[Sentiment]: Positive
###
[Review]: "I had a terrible experience at this place. The food was cold, and the staff was rude."
[Sentiment]: Negative
###
[Review]: "The ambiance and decor were charming, but the food quality was mediocre."
[Sentiment]: Neutral"""

# List of new restaurant reviews to predict sentiment
new_reviews = [
    "The atmosphere of this restaurant was pleasant, but the food was disappointing. I wouldn't recommend it.",
    "What a great experience! The food was amazing, and the staff was friendly and helpful.",
    "The restaurant is overpriced for the quality of the food, and the service is good.",
    "A cozy place with good food, but the music was too loud for my liking."
]

# Function to predict sentiment for multiple reviews
def predict_sentiment_for_reviews(reviews):
    for review in reviews:
        updated_prompt = prompt + f"\n[Review]: \"{review}\"\n[Sentiment]:"

        try:
            # Generate text based on the updated prompt using few-shot learning and EOS token ID
            results = few_shot_pipeline(updated_prompt, eos_token_id=eos_token_id)

            # Extract the sentiment label from the results for the new review
            if results and len(results) > 0:
                generated_text = results[0]["generated_text"].strip()

                # Use regex to capture the sentiment
                pattern = r"\[Sentiment\]:\s*(\w+)"
                matches = re.findall(pattern, generated_text)

                if matches:
                    last_sentiment = matches[-1].lower()  # Get the last match and convert to lowercase
                    print(f"Review: {review}")
                    print(f"Predicted Sentiment: {last_sentiment}\n")
                else:
                    print("Sentiment not found in the generated text.")
            else:
                print("Error: No results returned from the model.")

        except Exception as e:
            print(f"An error occurred for review '{review}': {e}")

# Run sentiment prediction for the list of reviews
predict_sentiment_for_reviews(new_reviews)
```
